# Preprocessing 3

```python
import pandas as pd
import numpy as np
df=pd.read_csv("aiddata-countries-only - aiddata-countries-only.csv")
df13=df[df.coalesced_purpose_name != 'Sectors not specified']
ls=df13['coalesced_purpose_name'].value_counts().head()
ls=pd.DataFrame(ls)
causes=list(ls.index)
causes
```

```python
df12=df[df.coalesced_purpose_name.isin(causes)]
df6=df12[["recipient","coalesced_purpose_name","commitment_amount_usd_constant"]]
df6.head()
```

```python
# df7=pd.DataFrame(df6.groupby(['recipient', 'coalesced_purpose_name']).size())
# df7=df7.reset_index()
# print(df7.columns)
# df7.rename(columns={ df7.columns[2]: "COUNTER" }, inplace = True)
df6['COUNTER'] =1
group_data = df6.groupby(["recipient","coalesced_purpose_name"])['COUNTER'].sum() #sum function
df7=pd.DataFrame(group_data)
df7
```

```python
newdf=df7.groupby(["recipient"]).sum().sort_values(by=['COUNTER'],ascending=False)
newdf=newdf.rename(columns={"COUNTER": "SUM"})
newdf.head()
```

```python
lol=newdf.join(df7,how="inner")
lol=lol.sort_values(by=['SUM','recipient'],ascending=[False,True])
lol
```

```python
lol=lol.reset_index()
lol=lol.set_index('recipient')
lol=lol.drop(["SUM"],axis=1)
#lol.to_csv("data3.csv")
lol
```
