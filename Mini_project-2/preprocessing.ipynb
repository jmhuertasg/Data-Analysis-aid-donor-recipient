```python
df=pd.read_csv("aiddata-countries-only - aiddata-countries-only.csv")
df.head()
```

```python
import pandas as pd
data = pd.read_csv('test.csv',encoding="utf-16", sep='\t')
data.head()
```

```python
data['country']=data['country'].fillna(data['country_recieved'])
data['year']=data['year'].fillna(data['year_recieved'])
d2=data[['country','donated','recieved','year']]
```

```python
d3=d2[['year','country','don_rec']]
d3.year=d3.year.astype(int)
d3
```

```python
d3=d3.reset_index()
d3=d3.set_index('year')
d3
#d3.to_csv('data3.csv')
```
